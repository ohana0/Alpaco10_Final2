```python
# PDF 파일 경로 설정
pdf_path = "[2조 뇌 영역, 위치별 손상 정보 찾는 전문 서적]adams-and-victors-principles-of-neurology-8th.pdf"
```

```python
from PyPDF2 import PdfReader

def make_data(pdf_path):
    # PDF 파일을 읽고 청크로 나누기
    reader = PdfReader(pdf_path)
    chunks = []  # 청크를 저장할 리스트

    for page_number, page in enumerate(reader.pages):
        # 각 페이지에서 텍스트 추출
        page_content = page.extract_text()
        if page_content:
            chunk = {
                'content': page_content,
                'metadata': {'page_number': page_number}  # 페이지 번호 추가
            }
            chunks.append(chunk)

    # 데이터베이스 연결 및 데이터 삽입
    conn = sqlite3.connect('pdf_database.db')
    cursor = conn.cursor()

    for chunk in chunks:
        page_number = chunk['metadata']['page_number']  # 페이지 번호 가져오기
        cursor.execute("INSERT INTO pdf_chunks (content, page_number) VALUES (?, ?)", 
                       (chunk['content'], page_number))

    conn.commit()
    return cursor
```

```python
import os
import sqlite3
from PyPDF2 import PdfReader
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from langchain.chains import RetrievalQA
from langchain.llms import Ollama
from langchain.vectorstores import FAISS
from langchain.embeddings import HuggingFaceEmbeddings

# PDF에서 데이터를 추출하고 SQLite DB에 저장하는 함수
def make_data(pdf_path):
    reader = PdfReader(pdf_path)
    chunks = []  # 청크를 저장할 리스트

    for page_number, page in enumerate(reader.pages):
        page_content = page.extract_text()  # 각 페이지에서 텍스트 추출
        if page_content:
            chunk = {
                'content': page_content,
                'metadata': {'page_number': page_number}  # 페이지 번호 추가
            }
            chunks.append(chunk)

    # 데이터베이스 연결 및 데이터 삽입
    conn = sqlite3.connect('pdf_database.db')
    cursor = conn.cursor()

    # 테이블이 없으면 생성
    cursor.execute('''CREATE TABLE IF NOT EXISTS pdf_chunks (
                        content TEXT,
                        page_number INTEGER)''')

    for chunk in chunks:
        page_number = chunk['metadata']['page_number']
        cursor.execute("INSERT INTO pdf_chunks (content, page_number) VALUES (?, ?)", 
                       (chunk['content'], page_number))

    conn.commit()
    conn.close()
    return chunks


# 데이터베이스 연결 확인 및 PDF 데이터 삽입
def load_pdf_data(pdf_path):
    # 데이터베이스가 존재하지 않으면 새로운 데이터를 삽입
    if not os.path.exists('pdf_database.db'):
        print("새 데이터베이스 생성 중...")
        make_data(pdf_path)
    else:
        conn = sqlite3.connect('pdf_database.db')
        cursor = conn.cursor()
        cursor.execute("SELECT content FROM pdf_chunks WHERE content LIKE ?", ('%' + pdf_path.split('/')[-1] + '%',))
        existing_data = cursor.fetchall()
        
        if not existing_data:
            print("파일이 없으므로 새로운 데이터 삽입 중...")
            make_data(pdf_path)
        else:
            print("PDF 파일이 이미 데이터베이스에 존재합니다.")
        conn.close()

# 임베딩 생성 및 FAISS 인덱스 생성
def create_vector_store():
    conn = sqlite3.connect('pdf_database.db')
    cursor = conn.cursor()
    
    # 데이터베이스에서 텍스트 추출
    cursor.execute("SELECT content FROM pdf_chunks")
    db_chunks = cursor.fetchall()
    texts = [chunk[0] for chunk in db_chunks]

    # HuggingFace 모델을 사용한 임베딩 생성
    embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
    
    # FAISS 벡터 저장소 생성
    vectorstore = FAISS.from_texts(texts, embeddings)
    conn.close()
    return vectorstore

# 라벨에 따른 설명을 query로 사용
queries = {
    0: "Background, gray color, non-tumor areas",
    1: "Necrotic tumor core (NCR) and non-enhancing tumor core (NET), low cell activity, treatment resistance",
    2: "Peritumoral edema, tumor invasion, critical for therapy planning and surgical resection",
    4: "Enhancing tumor, gadolinium enhancement, active tumor, breach of blood-brain barrier"
}

# 각 라벨에 대한 검색 수행
def search_label(label, vectorstore):
    query = queries.get(label)
    if query:
        # 벡터 저장소에서 유사도 검색
        results = vectorstore.similarity_search(query, k=3)  # k는 결과 개수
        return results
    else:
        print(f"라벨 {label}에 해당하는 query가 없습니다.")
        return []

def main():
    # PDF 파일 경로 설정
    pdf_path = "[2조 뇌 영역, 위치별 손상 정보 찾는 전문 서적]adams-and-victors-principles-of-neurology-8th.pdf"

    # PDF 데이터 로드 및 데이터베이스에 삽입
    load_pdf_data(pdf_path)

    # 벡터 저장소 생성
    vectorstore = create_vector_store()

    # 검색기 설정
    retriever = vectorstore.as_retriever(search_type="similarity", search_kwargs={"k": 2})

    # Ollama 모델 생성
    ollama_gemma2 = Ollama(model="gemma2")

    # QA 체인 생성
    qa = RetrievalQA.from_chain_type(llm=ollama_gemma2, chain_type="stuff", retriever=retriever)

    # 각 라벨에 대한 검색 및 요약 수행
    for label in queries:
        print(f"라벨 {label}에 대한 검색 결과:")
        search_results = search_label(label, vectorstore)
        
        # 검색 결과가 있으면 요약 수행
        if search_results:
            results_text = "\n".join([result.page_content for result in search_results])
            query = f"이 텍스트를 요약해주세요: {results_text}"
            
            # 사용자 쿼리로 응답 생성
            response = qa.invoke(query)
            print(f"요약 결과 {label}: {response}")
        else:
            print(f"라벨 {label}에 대한 검색 결과가 없습니다.")
        
        print("-" * 50)  # 구분선

if __name__ == "__main__":
    main()
```

```python
import os
import sqlite3
from PyPDF2 import PdfReader
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from langchain.chains import RetrievalQA
from langchain.llms import Ollama
from langchain.vectorstores import FAISS
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.schema import Document
from langchain.retrievers import BM25Retriever, EnsembleRetriever

# PDF에서 데이터를 추출하고 SQLite DB에 저장하는 함수
def make_data(pdf_path):
    reader = PdfReader(pdf_path)
    chunks = []  # 청크를 저장할 리스트

    for page_number, page in enumerate(reader.pages):
        page_content = page.extract_text()  # 각 페이지에서 텍스트 추출
        if page_content:
            chunk = {
                'content': page_content,
                'metadata': {'page_number': page_number}  # 페이지 번호 추가
            }
            chunks.append(chunk)

    # 데이터베이스 연결 및 데이터 삽입
    conn = sqlite3.connect('pdf_database.db')
    cursor = conn.cursor()

    # 테이블이 없으면 생성
    cursor.execute('''CREATE TABLE IF NOT EXISTS pdf_chunks (
                        content TEXT,
                        page_number INTEGER)''')

    for chunk in chunks:
        page_number = chunk['metadata']['page_number']
        cursor.execute("INSERT INTO pdf_chunks (content, page_number) VALUES (?, ?)", 
                       (chunk['content'], page_number))

    conn.commit()
    conn.close()
    return chunks


# 데이터베이스 연결 확인 및 PDF 데이터 삽입
def load_pdf_data(pdf_path):
    # 데이터베이스가 존재하지 않으면 새로운 데이터를 삽입
    if not os.path.exists('pdf_database.db'):
        print("새 데이터베이스 생성 중...")
        make_data(pdf_path)
    else:
        conn = sqlite3.connect('pdf_database.db')
        cursor = conn.cursor()
        cursor.execute("SELECT content FROM pdf_chunks WHERE content LIKE ?", ('%' + pdf_path.split('/')[-1] + '%',))
        existing_data = cursor.fetchall()
        
        if not existing_data:
            print("파일이 없으므로 새로운 데이터 삽입 중...")
            make_data(pdf_path)
        else:
            print("PDF 파일이 이미 데이터베이스에 존재합니다.")
        conn.close()

# 임베딩 생성 및 FAISS 인덱스 생성
def create_vector_store():
    conn = sqlite3.connect('pdf_database.db')
    cursor = conn.cursor()
    
    # 데이터베이스에서 텍스트 추출
    cursor.execute("SELECT content FROM pdf_chunks")
    db_chunks = cursor.fetchall()
    texts = [chunk[0] for chunk in db_chunks]

    # HuggingFace 모델을 사용한 임베딩 생성
    embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
    
    # FAISS 벡터 저장소 생성
    vectorstore = FAISS.from_texts(texts, embeddings)
    conn.close()
    return vectorstore

# 라벨에 따른 설명을 query로 사용
queries = {
    0: "Background, gray color, non-tumor areas",
    1: "Necrotic tumor core (NCR) and non-enhancing tumor core (NET), low cell activity, treatment resistance",
    2: "Peritumoral edema, tumor invasion, critical for therapy planning and surgical resection",
    4: "Enhancing tumor, gadolinium enhancement, active tumor, breach of blood-brain barrier"
}

# 각 라벨에 대한 검색 수행
def search_label(label, ensemble_retriever):
    query = queries.get(label)
    if query:
        # 앙상블 리트리버를 사용하여 유사도 검색
        results = ensemble_retriever.get_relevant_documents(query)  # BM25 + FAISS 리트리버를 통해 검색
        return results
    else:
        print(f"라벨 {label}에 해당하는 query가 없습니다.")
        return []

def main():
    # PDF 파일 경로 설정
    pdf_path = "[2조 뇌 영역, 위치별 손상 정보 찾는 전문 서적]adams-and-victors-principles-of-neurology-8th.pdf"

    # PDF 데이터 로드 및 데이터베이스에 삽입
    load_pdf_data(pdf_path)

    # 벡터 저장소 생성
    vectorstore = create_vector_store()

    # 벡터 저장소 기반 FAISS 리트리버 생성
    faiss_retriever = vectorstore.as_retriever(search_kwargs={"k": 2})

    # BM25 리트리버 생성
    conn = sqlite3.connect('pdf_database.db')
    cursor = conn.cursor()
    cursor.execute("SELECT content FROM pdf_chunks")
    db_chunks = cursor.fetchall()
    texts = [chunk[0] for chunk in db_chunks]
    bm25_retriever = BM25Retriever.from_documents([Document(page_content=text) for text in texts])
    bm25_retriever.k = 2

    # EnsembleRetriever 초기화 (BM25 + FAISS 리트리버)
    ensemble_retriever = EnsembleRetriever(
        retrievers=[bm25_retriever, faiss_retriever], weights=[0.5, 0.5]
    )

    # 각 라벨에 대한 검색 및 요약 수행
    for label in queries:
        print(f"라벨 {label}에 대한 검색 결과:")
        search_results = search_label(label, ensemble_retriever)
        
        # 검색 결과가 있으면 요약 수행
        if search_results:
            results_text = "\n".join([result.page_content for result in search_results])
            query = f"이 텍스트를 요약해주세요: {results_text}"
            
            # 사용자 쿼리로 응답 생성
            # Ollama 모델 사용 (여기서는 Ollama 대신 다른 모델을 사용할 수 있음)
            ollama_gemma2 = Ollama(model="gemma2")
            qa = RetrievalQA.from_chain_type(llm=ollama_gemma2, chain_type="stuff", retriever=ensemble_retriever)
            response = qa.invoke(query)
            print(f"요약 결과 {label}: {response}")
        else:
            print(f"라벨 {label}에 대한 검색 결과가 없습니다.")
        
        print("-" * 50)  # 구분선

if __name__ == "__main__":
    main()
```

```python
import os
import sqlite3
from PyPDF2 import PdfReader
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from langchain.chains import RetrievalQA
from langchain.llms import Ollama
from langchain.schema import Document
from langchain.retrievers import BM25Retriever, EnsembleRetriever

# PDF에서 데이터를 추출하고 SQLite DB에 저장하는 함수
def make_data(pdf_path):
    reader = PdfReader(pdf_path)
    chunks = []

    for page_number, page in enumerate(reader.pages):
        page_content = page.extract_text()
        if page_content:
            chunk = {
                'content': page_content,
                'metadata': {'page_number': page_number + 1}  # 페이지 번호(1부터 시작)
            }
            chunks.append(chunk)

    conn = sqlite3.connect('pdf_database.db')
    cursor = conn.cursor()

    cursor.execute('''CREATE TABLE IF NOT EXISTS pdf_chunks (
                        content TEXT,
                        page_number INTEGER)''')

    for chunk in chunks:
        page_number = chunk['metadata']['page_number']
        cursor.execute("INSERT INTO pdf_chunks (content, page_number) VALUES (?, ?)", 
                       (chunk['content'], page_number))

    conn.commit()
    conn.close()
    return chunks

# 데이터베이스 연결 확인 및 PDF 데이터 삽입
def load_pdf_data(pdf_path):
    if not os.path.exists('pdf_database.db'):
        print("새 데이터베이스 생성 중...")
        make_data(pdf_path)
    else:
        conn = sqlite3.connect('pdf_database.db')
        cursor = conn.cursor()
        cursor.execute("SELECT content FROM pdf_chunks")
        existing_data = cursor.fetchall()
        
        if not existing_data:
            print("파일이 없으므로 새로운 데이터 삽입 중...")
            make_data(pdf_path)
        else:
            print("PDF 파일이 이미 데이터베이스에 존재합니다.")
        conn.close()

# 임베딩 생성 및 FAISS 인덱스 생성
def create_vector_store():
    conn = sqlite3.connect('pdf_database.db')
    cursor = conn.cursor()
    
    cursor.execute("SELECT content, page_number FROM pdf_chunks")
    db_chunks = cursor.fetchall()
    texts = [chunk[0] for chunk in db_chunks]
    metadata = [{'page_number': chunk[1]} for chunk in db_chunks]

    embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
    vectorstore = FAISS.from_texts(texts, embeddings, metadatas=metadata)
    conn.close()
    return vectorstore

# 각 라벨에 대한 검색 수행
def search_label(label, retriever, query):
    results = retriever.get_relevant_documents(query)
    if results:
        # 페이지 번호와 텍스트 수집
        summarized_texts = []
        for result in results:
            content = result.page_content
            page_number = result.metadata['page_number']
            summarized_texts.append(f"(Page {page_number}) {content[:500]}...")
        return summarized_texts
    return []

def main():
    pdf_path = "[2조 뇌 영역, 위치별 손상 정보 찾는 전문 서적]adams-and-victors-principles-of-neurology-8th.pdf"
    load_pdf_data(pdf_path)

    vectorstore = create_vector_store()
    faiss_retriever = vectorstore.as_retriever(search_kwargs={"k": 3})

    # 라벨에 해당하는 질환명 추가
    label_names = {
        0: "Non-tumor areas",
        1: "Necrotic tumor core and non-enhancing tumor core",
        2: "Peritumoral edema and tumor invasion",
        4: "Enhancing tumor with gadolinium enhancement"
    }

    queries = {
        0: "Background, gray color, non-tumor areas",
        1: "Necrotic tumor core (NCR) and non-enhancing tumor core (NET), low cell activity, treatment resistance",
        2: "Peritumoral edema, tumor invasion, critical for therapy planning and surgical resection",
        4: "Enhancing tumor, gadolinium enhancement, active tumor, breach of blood-brain barrier"
    }

    for label, query in queries.items():
        print(f"\n{label_names[label]}에 대한 검색 결과:")
        results = search_label(label, faiss_retriever, query)
        
        # 검색된 결과 텍스트 요약
        combined_text = " ".join(results)
        ollama_gemma2 = Ollama(model="gemma2")
        qa = RetrievalQA.from_chain_type(llm=ollama_gemma2, chain_type="stuff", retriever=faiss_retriever)
        
        # 요약을 1000자로 요청하고 치료 계획과 주의사항도 포함
        summary = qa.invoke(f"Summarize the following text in 1000 characters including treatment plan and precautions: {combined_text}")
        
        print(f"요약 결과 ({label_names[label]}):\n{summary}")
        print("-" * 50)

if __name__ == "__main__":
    main()
```

```python
import os
import sqlite3
from PyPDF2 import PdfReader
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from langchain.chains import RetrievalQA
from langchain.llms import Ollama
from langchain.schema import Document
from langchain.retrievers import BM25Retriever, EnsembleRetriever

# PDF에서 데이터를 추출하고 SQLite DB에 저장하는 함수
def make_data(pdf_path):
    reader = PdfReader(pdf_path)
    chunks = []  # 청크를 저장할 리스트

    for page_number, page in enumerate(reader.pages):
        page_content = page.extract_text()  # 각 페이지에서 텍스트 추출
        if page_content:
            chunk = {
                'content': page_content,
                'metadata': {'page_number': page_number}  # 페이지 번호 추가
            }
            chunks.append(chunk)

    # 데이터베이스 연결 및 데이터 삽입
    conn = sqlite3.connect('pdf_database.db')
    cursor = conn.cursor()

    # 테이블이 없으면 생성
    cursor.execute('''CREATE TABLE IF NOT EXISTS pdf_chunks (
                        content TEXT,
                        page_number INTEGER)''')

    for chunk in chunks:
        page_number = chunk['metadata']['page_number']
        cursor.execute("INSERT INTO pdf_chunks (content, page_number) VALUES (?, ?)", 
                       (chunk['content'], page_number))

    conn.commit()
    conn.close()
    return chunks


# 데이터베이스 연결 확인 및 PDF 데이터 삽입
def load_pdf_data(pdf_path):
    # 데이터베이스가 존재하지 않으면 새로운 데이터를 삽입
    if not os.path.exists('pdf_database.db'):
        print("새 데이터베이스 생성 중...")
        make_data(pdf_path)
    else:
        conn = sqlite3.connect('pdf_database.db')
        cursor = conn.cursor()
        cursor.execute("SELECT * FROM pdf_chunks")
        existing_data = cursor.fetchall()
        
        if not existing_data:
            print("파일이 없으므로 새로운 데이터 삽입 중...")
            make_data(pdf_path)
        else:
            print("PDF 데이터가 이미 데이터베이스에 존재합니다.")
        conn.close()

# 임베딩 생성 및 FAISS 인덱스 생성
def create_vector_store():
    conn = sqlite3.connect('pdf_database.db')
    cursor = conn.cursor()
    
    # 데이터베이스에서 텍스트 추출
    cursor.execute("SELECT content FROM pdf_chunks")
    db_chunks = cursor.fetchall()
    texts = [chunk[0] for chunk in db_chunks]

    # HuggingFace 모델을 사용한 임베딩 생성
    embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
    
    # FAISS 벡터 저장소 생성
    vectorstore = FAISS.from_texts(texts, embeddings)
    conn.close()
    return vectorstore

# 라벨에 따른 설명을 query로 사용
queries = {
    0: "Background, gray color, non-tumor areas",
    1: "Necrotic tumor core (NCR) and non-enhancing tumor core (NET), low cell activity, treatment resistance",
    2: "Peritumoral edema, tumor invasion, critical for therapy planning and surgical resection",
    4: "Enhancing tumor, gadolinium enhancement, active tumor, breach of blood-brain barrier"
}

# 출력 포맷을 보기 좋게 정리하는 함수
def format_output(qa_responses, queries):
    if not isinstance(queries, dict) or not isinstance(qa_responses, dict):
        raise ValueError("Both 'queries' and 'qa_responses' must be dictionaries.")
    
    formatted_response = "### Results:\n"
    for label, query in queries.items():
        formatted_response += f"\n\n### Query (Label {label}): \n{query}\n"
        response = qa_responses.get(label, "응답 없음")
        formatted_response += f"**Result:**\n{response}"
    return formatted_response

def main():
    pdf_path = "[2조 뇌 영역, 위치별 손상 정보 찾는 전문 서적]adams-and-victors-principles-of-neurology-8th.pdf"

    # PDF 데이터 로드 및 데이터베이스에 삽입
    load_pdf_data(pdf_path)

    # 벡터 저장소 생성
    vectorstore = create_vector_store()
    faiss_retriever = vectorstore.as_retriever(search_kwargs={"k": 2})

    # BM25 리트리버 생성
    conn = sqlite3.connect('pdf_database.db')
    cursor = conn.cursor()
    cursor.execute("SELECT content FROM pdf_chunks")
    texts = [chunk[0] for chunk in cursor.fetchall()]
    bm25_retriever = BM25Retriever.from_documents([Document(page_content=text) for text in texts])
    bm25_retriever.k = 2

    # EnsembleRetriever 초기화
    ensemble_retriever = EnsembleRetriever(
        retrievers=[bm25_retriever, faiss_retriever], weights=[0.5, 0.5]
    )

    # Ollama 모델 초기화
    ollama_gemma2 = Ollama(model="gemma2")
    qa = RetrievalQA.from_chain_type(llm=ollama_gemma2, chain_type="stuff", retriever=ensemble_retriever)

    # 각 라벨에 대해 검색 및 처리
    qa_responses = {}
    for label, query in queries.items():
        print(f"라벨 {label} 처리 중...")
        response = qa.invoke(f"Summarize the following medical information: {query}")
        qa_responses[label] = response.get("result", "응답 없음")

    # 포맷 함수로 출력
    formatted_response = format_output(qa_responses, queries)
    print(formatted_response)

if __name__ == "__main__":
    main()
```
